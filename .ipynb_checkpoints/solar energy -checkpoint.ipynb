```python
# Defining the Input variable for the Selection of important features

X= solar.drop(['Power Generated in MW','Solar', 'Month ','Day'], axis = 1) # high +ve correlation with Power generated, to avoid overfitting 


# Y is output 
y = solar['Power Generated in MW'].astype('int')


# Splitting the dataset into train and test data

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)


# Scaling the Training Data / # Z-score the features

#sc=StandardScaler()
#X= sc.fit(X_train, y_train.astype('int'))

from sklearn import preprocessing
mm_scaler = preprocessing.MinMaxScaler()
X_train_minmax = mm_scaler.fit_transform(X_train.astype('int'))#
mm_scaler.transform(X_test.astype('int'))#
```

```python
# Feature Engineering with Random Forest Classifier 


# Fitting the model selection with the RFE 

sel = SelectFromModel(RandomForestRegressor(n_estimators = 100))
sel.fit(X_train, y_train)
```

```python
sel.get_support()
```

```python
# list and count the selected features

selected_feat= X_train.columns[(sel.get_support())]
selected_feat
```

```python
# plot the features in the order of increasing importantance

plt.figure(figsize=(10,7))
plt.title('Features Importance with RFR')
feat_importances =pd.Series(sel.estimator_.feature_importances_, index = X_train.columns)
feat_importances.nlargest().plot(kind='barh');
```

```python
# identify a relationship between the two variables;  Global_active_power & Global_intensity
sns.set_style('white')  # Style of figure to be plotted 
sns.set_style('ticks')
plt.figure(figsize = (8,6))  # Defining the figsize 
sns.regplot(x='Cloud Cover Percentage',y='Power Generated in MW',color='gold', data=solar)
plt.title('Relationship between Cloud Cover Percentage and Power Generated in MW', fontsize = 15)  # Title of graph 
```

```python
# Defining the new X and Y variables given the RFE outcomes.

X1 = solar[['Cloud Cover Percentage']].values 
y1 = solar['Power Generated in MW'].values
```

```python
# Splitting the dataset into train and test data

X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.20, random_state=42)


# Scaling the Training Data / # Z-score the features


mm_scaler = preprocessing.MinMaxScaler()
X_train_minmax = mm_scaler.fit_transform(X1_train.astype('int'))#
mm_scaler.transform(X1_test.astype('int'))#
```

```python
# Training the model

reg = LinearRegression()  
reg.fit(X1_train, y1_train)

```

```python
# Testing the model with our testing dataset.
y_pred=reg.predict(X1_test)
```

```python
# Checking the intercept 
print("Intercept: ", reg.intercept_)
```

```python
# Checking the coefficients 
Coef = reg.coef_
(Coef)
```

```python
# Score Model
print("R^2 Accuracy: ",reg.score(X1_test, y1_test))
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y1_test, y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(y1_test, y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y1_test, y_pred)))
```

```python
#Generated Predictions

#Plot Our Actual and Predicted Values

plt.plot(X1_test, y1_test, 'o', color='red');
plt.plot(X1_test,y_pred,color='blue')
plt.title("Actuals vs Regression Line")

X1 = solar[['Cloud Cover Percentage']].values 
y1 = solar['Power Generated in MW'].values
```

```python
## for machine learning
from sklearn import model_selection, preprocessing, feature_selection, ensemble, linear_model, metrics, decomposition

model = linear_model.LinearRegression()

scores = []
cv = model_selection.KFold(n_splits=5, shuffle=True)
fig = plt.figure()
i = 1
for train, test in cv.split(X1_train, y1_train):
    prediction = model.fit(X1_train[train],
                 y1_train[train]).predict(X1_train[test])
    true = y1_train[test]
    score = metrics.r2_score(true, prediction)
    scores.append(score)
    plt.scatter(prediction, true, lw=2, alpha=0.3, 
                label='Fold %d (R2 = %0.2f)' % (i,score))
    i = i+1
plt.plot([min(y1_train),max(y1_train)], [min(y1_train),max(y1_train)], 
         linestyle='--', lw=2, color='black')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('K-Fold Validation')
plt.legend()
plt.show()
```

```python
## Kpi
print("R2 (explained variance):", round(metrics.r2_score(y1_test, y_pred), 2))
print("Mean Absolute Perc Error (Σ(|y-pred|/y)/n):", round(np.mean(np.abs((y1_test-y_pred)/y_pred)), 2))
print("Mean Absolute Error (Σ|yi-pred|/n):", "{:,.0f}".format(metrics.mean_absolute_error(y1_test, y_pred)))
print("Root Mean Squared Error (sqrt(Σ(y-pred)^2/n)):", "{:,.0f}".format(np.sqrt(metrics.mean_squared_error(y1_test, y_pred))))## residuals


residuals = y1_test - y_pred
max_error = max(residuals) if abs(max(residuals)) > abs(min(residuals)) else min(residuals)
max_idx = list(residuals).index(max(residuals)) if abs(max(residuals)) > abs(min(residuals)) else list(residuals).index(min(residuals))
max_true, max_pred = y1_test[max_idx], y_pred[max_idx]
print("Max Error:", "{:,.0f}".format(max_error))
```

```python
## Plot predicted vs true
fig, ax = plt.subplots(nrows=1, ncols=2)
from statsmodels.graphics.api import abline_plot
ax[0].scatter(y_pred, y1_test, color="blue")
abline_plot(intercept=0, slope=1, color="red", ax=ax[0])
ax[0].vlines(x=max_pred, ymin=max_true, ymax=max_true-max_error, color='green', linestyle='--', alpha=0.9, label="max error")
ax[0].grid(True)
ax[0].set(xlabel="Predicted", ylabel="True", title="Predicted vs True")
ax[0].legend()
    
## Plot predicted vs residuals
ax[1].scatter(y_pred, residuals, color="red")
ax[1].vlines(x=max_pred, ymin=0, ymax=max_error, color='black', linestyle='--', alpha=0.9, label="max error")
ax[1].grid(True)
ax[1].set(xlabel="Predicted", ylabel="Residuals", title="Predicted vs Residuals")
ax[1].hlines(y=0, xmin=np.min(y_pred), xmax=np.max(y_pred))
ax[1].legend()
plt.show()
```
